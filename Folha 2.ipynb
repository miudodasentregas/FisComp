```python
import pylab as p
import numpy as np
import math


data = np.loadtxt("sunspots.txt",float)
x = data[:,0]
y = data[:,1]
yav = []
for i in range(len(y) ):
    yav.append((1/11)*sum(y[i-5:i+6]))
            
p.xlabel('anos')
p.ylabel('sunspots')
p.xlim(0,1000)
p.plot(x,y)
p.plot(x,yav)
p.show()
```

```python
import pylab as p
import numpy as np
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)

matriz = np.zeros((100,100))
for i in range(100):
    for j in range(100):
        z = np.zeros(2)
        a = 0
        for k in range(1000):
            z = np.array([z[0]**2-z[1]**2,2*z[0]*z[1]]) + np.array([x[j],y[i]])
            if (z[0]**2 + z[1]**2) > 4:
                a = k
                break
            matriz[i][j] = k
p.imshow(matriz, origin='lower',extent=[-2,2,-2,2])
p.jet()
p.show()
```
